```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
df = pd.read_csv('Diwali Sales Data.csv', encoding='unicode_escape')
# to avoid encoding error use 'unicode_escape'
## get an overview of the datset
df.head(10)
```

```python
## get the number of rows and columns in the  dataset
df.shape
```

```python
## check the data type of each column
df.info()
```

```python
## describe the data set to better understand the variance of the columns and for initial analysis get
## count, max, min, top values from the dataset
df.describe(include = 'all')
```

```python
## drop thr unrelated/ blank columns
df.drop(['Status', 'unnamed1'], axis = 1, inplace = True)
```

```python
## check null valus in the columns

pd.isnull(df).sum()
```

```python
## delete the null rows

df.dropna(inplace=True)
```

```python
df.shape
```

```python
df['Amount'] =  df['Amount'].astype('int')
```

# Exploratory Data Analysis

**Gender**

```python
ax = sns.countplot(x = 'Gender', data = df, palette='Set1')
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_based_on_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x = 'Gender', y='Amount', data= sales_based_on_gender, palette='Set1')
```

# From the above graphs, we see that most of the buyers are Female and has more revenue contribution than the men.

**Age**

```python
ax = sns.countplot(x= 'Age Group', data = df, hue='Gender')
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
## Total Amount vs AGe group
sales_based_on_age_group = df.groupby(['Age Group'], as_index = False)['Amount'].sum().sort_values(by='Amount', ascending=False)
sns.barplot(x= 'Age Group', y = 'Amount', data = sales_based_on_age_group, palette='Set2')
```

# From the above graphs, we see that most of the buyers are of age 26-35 group and the females constitute most of this group.

**State**

```python
## total number of orders from the top 10 states
orders_based_on_state = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize' : (18, 5)})
sns.barplot(data = orders_based_on_state, x= 'State', y ='Orders', palette='Set2')
```

```python
## total amount or sales earned per top 10 states
sales_based_on_state = df.groupby(['State'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize' : (18, 5)})
sns.barplot(data = sales_based_on_state, x= 'State', y ='Amount', palette='Set2')
```

# From the graphs, we cam see that most of the orders placed and total sales/amount earned are from the Uttar Pradesh, Maharashta and Karnataka in the order of descreasing.

**Marital Status**

```python
ax = sns.countplot(x= 'Marital_Status', data = df, palette='Set1')
sns.set(rc={'figure.figsize' : (7, 5)})
for bars in ax.containers:
  ax.bar_label(bars)
```

```python
## total amount or sales earned based on Marital Status
sales_based_on_Marital_status = df.groupby(['Marital_Status', 'Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize' : (6, 5)})
sns.barplot(data = sales_based_on_Marital_status, x= 'Marital_Status', y ='Amount', hue='Gender')
```

# Most of the buyers are married women

**Occupation**


```python

sns.set(rc={'figure.figsize' : (20, 5)})
ax = sns.countplot(x= 'Occupation', data = df, palette='Set1')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
## total amount or sales earned based on Occupation
sales_based_on_Occupation = df.groupby(['Occupation', 'Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize' : (20, 5)})
sns.barplot(data = sales_based_on_Occupation, x= 'Occupation', y ='Amount', palette='Set2')
```

# Most of the buyers are from IT , Aviation and Healthcare sector

**Product Category**

```python
sns.set(rc={'figure.figsize' : (30, 5)})
ax = sns.countplot(x= 'Product_Category', data = df, palette='Set1')

for bars in ax.containers:
  ax.bar_label(bars)
```

```python
## total amount or sales earned based on Product_Category
sales_based_on_Product_Category = df.groupby(['Product_Category', 'Gender'], as_index= False)['Amount'].sum().sort_values(by='Amount', ascending = False).head(10)
sns.set(rc={'figure.figsize' : (20, 5)})
sns.barplot(data = sales_based_on_Product_Category, x= 'Product_Category', y ='Amount', palette='Set2')
```

# Most of the sold products belong to Food, Foorwear and Electronics category.

```python
sale_per_state = df.groupby(['Product_ID'], as_index= False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set(rc={'figure.figsize' : (20, 5)})
sns.barplot(data = sale_per_state, x= 'Product_ID', y ='Orders', palette='Set2')
```
